```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('matches.csv')
```

```python
df.head()
```

```python
def insert_column(row):
    if pd.isnull(row['winner_runs']):
        return 'field1st'
    else:
        return 'bat1st'

# Inserting new column based on column 'A'
df['winner_were'] = df.apply(insert_column, axis=1)

print(df)
```

```python
df.head()
```

```python
city = df.pivot(columns='city', values='winner_were')

city
```

```python
city.head()
```

```python
toss = df.pivot(columns='city', values='toss_decision')
```

```python
toss
```

```python
df['same_winner'] = df['toss_winner'] == df['winner']

df.head()
```

```python
same_winner = df.pivot(columns='city', values='same_winner')
```

```python
same_winner.head()
```

```python
bat_df = pd.read_csv('batting_summary.csv')
```

```python
bat_df.head()
```

```python
fours = bat_df.pivot_table(columns='Team_Innings', values='4s', aggfunc = 'sum')

fours.head()
```

```python
sixes = bat_df.pivot_table(columns='Team_Innings', values='6s', aggfunc = 'sum')

sixes.head()
```

```python
bat_df['Strike_Rate'] = pd.to_numeric(bat_df['Strike_Rate'], errors='coerce')

# Create pivot table
strike_rate = bat_df.pivot_table(columns='Team_Innings', values='Strike_Rate', aggfunc='mean')

strike_rate.head()
```

```python
count_greater_than_100 = bat_df[bat_df['Runs'] > 100].groupby('Team_Innings').size()
```

```python
count_greater_than_100
```

```python
count_greater_than_100 = bat_df[bat_df['Runs'] > 100].groupby('Team_Innings').size()
count_between_50_and_100 = bat_df[(bat_df['Runs'] >= 50) & (bat_df['Runs'] <= 100)].groupby('Team_Innings').size()


count_df = pd.DataFrame({
    '100s': count_greater_than_100,
    '50s': count_between_50_and_100
}).fillna(0)
```

```python
count_df
```

```python
player_runs = bat_df.groupby(['Team_Innings', 'Batsman_Name'])['Runs'].sum().reset_index()

player_runs_sorted = player_runs.sort_values(by='Runs', ascending=False)
```

```python
player_runs_sorted
```

```python
player_runs
```

```python
player_sr = bat_df.groupby(['Team_Innings', 'Batsman_Name'])['Strike_Rate'].mean().reset_index()

player_sr_sorted = player_sr.sort_values(by='Strike_Rate', ascending=False)
```

```python
player_sr_sorted 
```

```python
player_runs_temp = bat_df.groupby(['Team_Innings', 'Batsman_Name'])['Runs'].sum().reset_index()

player_runs_temp= player_runs.sort_values(by='Runs', ascending=False)
```

```python
player_runs_temp
```

```python
Top_Batsman= bat_df.groupby(['Batsman_Name', 'Team_Innings']).agg(
    Total_Runs=('Runs', 'sum'),
    Total_4s=('4s', 'sum'),
    Total_6s=('6s', 'sum'),
    Fifties=('Runs', lambda x: ((x >= 50) & (x <= 100)).sum()),
   Hundreds=('Runs', lambda x: (x > 100).sum()),
    Strike_rate=('Strike_Rate', 'mean')
).reset_index()

Top_Batsman =Top_Batsman.sort_values(by='Total_Runs', ascending=False)
```

```python
Top_Batsman 
```

```python
Team_Batting= bat_df.groupby(['Team_Innings']).agg(
    Total_Runs=('Runs', 'sum'),
    Total_4s=('4s', 'sum'),
    Total_6s=('6s', 'sum'),
    Fifties=('Runs', lambda x: ((x >= 50) & (x <= 100)).sum()),
   Hundreds=('Runs', lambda x: (x > 100).sum()),
    Strike_rate=('Strike_Rate', 'mean')
).reset_index()

Team_Batting = Team_Batting.sort_values(by='Total_Runs', ascending=False)
```

```python
Team_Batting 
```

```python
Team_Batting.to_csv('Team_Batting.csv', index=False)
```

```python
bowl_df = pd.read_csv('bowling_summary.csv')
```

```python
bowl_df.head()
```

```python
wickets = bowl_df.pivot_table(columns='Bowling_Team', values='Wickets', aggfunc = 'sum')

wickets.head()
```

```python
maidens = bowl_df.pivot_table(columns='Bowling_Team', values='Maidens', aggfunc = 'sum')

maidens.head()
```

```python
bowl_df['Economy'] = pd.to_numeric(bowl_df['Economy'], errors='coerce')

# Create pivot table
economy = bowl_df.pivot_table(columns='Bowling_Team', values='Economy', aggfunc='mean')

economy.head()
```

```python
player_wickets = bowl_df.groupby(['Bowling_Team', 'Bowler_Name'])['Wickets'].sum().reset_index()

player_wickets_sorted = player_wickets.sort_values(by='Wickets', ascending=False)
```

```python
player_wickets_sorted
```

```python
player_economy = bowl_df.groupby(['Bowling_Team', 'Bowler_Name'])['Economy'].mean().reset_index()

player_economy_sorted = player_economy.sort_values(by='Economy', ascending=True)
```

```python
player_economy_sorted
```

```python
Top_Bowlers= bowl_df.groupby(['Bowler_Name', 'Bowling_Team']).agg(
      Total_Over=('Overs', 'sum'),
    Total_Wickets=('Wickets', 'sum'),
    Total_Maidens=('Maidens', 'sum'),
   Five_Wickets_Haul=('Wickets', lambda x: (x > 5).sum()),
    Strike_rate=('Economy', 'mean')
).reset_index()

Top_Bowlers =Top_Bowlers.sort_values(by='Total_Wickets', ascending=False)
```

```python
Top_Bowlers
```

```python
Team_Bowling= bowl_df.groupby(['Bowling_Team']).agg(
      Total_Over=('Overs', 'sum'),
    Total_Wickets=('Wickets', 'sum'),
    Total_Maidens=('Maidens', 'sum'),
   Five_Wickets_Haul=('Wickets', lambda x: (x > 5).sum()),
    Strike_rate=('Economy', 'mean')
).reset_index()

Team_Bowling =Team_Bowling.sort_values(by='Total_Wickets', ascending=False)
```
